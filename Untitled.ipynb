```python
import pdbParser
import glob
import pdb
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt

plt.rcParams["figure.figsize"] = (20,20)


files = glob.glob('*nspose1.pdb')
print (files)
pdbs = []
#read in files 
for file in files:
    pdbs.append(pdbParser.pdbParser(file))
```

```python
#3d graph
startChain = pdbs[0].getChains()[0]
startAtoms = startChain.getAtoms()
endChain = pdbs[-1].getChains()[0]
endAtoms = endChain.getAtoms()


posArray = endChain.getPosArray()
xPos = posArray[:,0]
yPos = posArray[:,1]
zPos = posArray[:,2]
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(xPos, yPos, zPos, c='r', marker='o', s=300)

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
#Find max distance
distList = []
for key in startAtoms:
    distList.append(startAtoms[key].getDist(endAtoms[key]))
print(distList)

#color points based on distance from start
fig2 = plt.figure()
ax2 = fig2.add_subplot(111, projection='3d')
ax2.scatter(xPos, yPos, zPos, c= distList, marker='o', s=300)

ax2.set_xlabel('X Label')
ax2.set_ylabel('Y Label')
ax2.set_zlabel('Z Label')
plt.show()

```
